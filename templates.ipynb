Display the dataset 'ENERGY CONSUMPTION PER REGION'

```python
sns.barplot(data=data, y='Energy Consump.', x='Region')
plt.xticks(rotation=90)
plt.title('ENERGY CONSUMPTION PER REGION')
plt.ylabel('Energy Consumption')
plt.show()
```

###SunSaharan Africa has the highest Energy Consumption followed by South Asia. While Middle East and Noth Africa has the lowest Engery
consumption

```python
sns.barplot(data=data, y='Energy Consump.', x='Region', hue='Income Group')
plt.xticks(rotation=90)
plt.ylabel('Energy Consumption')
plt.title('Energy Consumption per Income Group')
```

##Sub Saharan Africa has highest Low income energy consumers, Latin America & Caribean has lowest Low income energy consumers.
North America has only High Income energy consumers
East Asia & Pacific region has highest Lower middle income who consume energy
Most energy consumers are Lower middle income group in 5 regions
