Problem Statement: Analysing Renewable Energy Consumption Patterns Across Countries
------------------=================================================================
Background:
-----------
Renewable energy consumption plays a crucial role in sustainable development and addressing climate change. Understanding the trends, patterns, and disparities in renewable energy consumption across countries can provide valuable insights for policymakers, energy planners, and researchers to promote renewable energy adoption and achieve global sustainability goals.
-------------------------------------------------------------------
Objective:
-----------
The objective of this analysis is to explore and analyse renewable energy consumption data across countries to identify trends, patterns, and variations in renewable energy usage. By examining key indicators such as country income group, region, and energy consumption levels over time, we aim to assess the progress, challenges, and opportunities for advancing renewable energy adoption on a global scale.
--------------------------------------------------------------------------------------------------


Datasets:
----------
The dataset comprises a table with the following columns:
Country Code: Unique code or identifier for each country.
Country Name: Name of the country.
Income Group: Economic classification of the country based on income level (e.g., low-income, middle-income, high-income).
Indicator Code: Code identifying the renewable energy consumption indicator.
Indicator Name: Description or name of the renewable energy consumption indicator.
Region: Geographic region to which the country belongs.
Year: Year of observation or data collection.
Energy Consump.: Renewable energy consumption level or value.

Analysis Tasks:
Global Renewable Energy Consumption Analysis:
Analyse overall trends and patterns in renewable energy consumption across countries and regions.
Identify countries and regions with high and low levels of renewable energy consumption.
Assess the growth and progress of renewable energy adoption over time.
Income Group Analysis:
-----------------------
Examine renewable energy consumption trends based on country income groups (e.g., low-income, middle-income, high-income).
Identify disparities and variations in renewable energy consumption levels among different income groups.
Assess the impact of economic factors on renewable energy adoption.
Regional Analysis:
------------------
Explore regional differences in renewable energy consumption patterns and trends.
Identify regions with significant renewable energy consumption growth or stagnation.
Assess regional policies and initiatives promoting renewable energy adoption.
Longitudinal Analysis:
----------------------
Conduct longitudinal analysis to examine changes in renewable energy consumption levels over time.
Identify factors influencing trends in renewable energy consumption and potential drivers of change.
Deliverables:
------------------------------------------------------------------

Data preprocessing and cleaning scripts for dataset preparation.
Exploratory data analysis report summarising key findings, visualisations, and insights.
Global renewable energy consumption analysis results highlighting trends and patterns across countries and regions.
Income group and regional analysis outcomes, including disparities and variations in renewable energy consumption levels.
Longitudinal analysis findings detailing changes in renewable energy consumption trends over time and potential influencing factors.
========================================================================================================================================

Conclusion:
By analysing renewable energy consumption data across countries and regions, this project aims to provide valuable insights into global renewable energy trends, patterns, and disparities. These insights can inform policy making, energy planning, and investment decisions aimed at promoting renewable energy adoption, enhancing energy security, and mitigating climate change impacts on a global scale.
